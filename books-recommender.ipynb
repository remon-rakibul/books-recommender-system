```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
books = books[
    ['ISBN',
    'Book-Title',
    'Book-Author',
    'Year-Of-Publication',
    'Publisher',
    'Image-URL-L']
]
```

```python
books.head()
```

```python
books.rename(columns={
    'ISBN': 'isbn',
    'Book-Title': 'title',
    'Book-Author': 'author',
    'Year-Of-Publication': 'year',
    'Publisher': 'publisher',
    'Image-URL-L': 'img_url'}, inplace=True)
```

```python
users = pd.read_csv('data/BX-Users.csv', sep=';', on_bad_lines='warn', encoding='latin-1')
```

```python
users.head()
```

```python
users.shape
```

```python
ratings = pd.read_csv('data/BX-Book-Ratings.csv', sep=';', on_bad_lines='warn', encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
print(books.shape)
print(users.shape)
print(ratings.shape)
```

```python
ratings.columns
```

```python
ratings.rename(columns={
    'User-ID': 'user_id',
    'ISBN': 'isbn',
    'Book-Rating': 'rating'}, inplace=True)
```

```python
ratings.head()
```

```python
ratings['user_id'].value_counts()
```

```python
ratings['user_id'].unique().shape
```

```python
x = ratings['user_id'].value_counts() > 200
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
books.head()
```

```python
ratings_with_books = ratings.merge(books, on='isbn')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
num_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
num_rating.head()
```

```python
num_rating.shape
```

```python
num_rating.rename(columns={'rating': 'num_of_rating'}, inplace=True)
```

```python
num_rating.head()
```

```python
ratings_with_books.head()
```

```python
final_rating = ratings_with_books.merge(num_rating, on='title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['num_of_rating'] >= 50]
```

```python
final_rating.shape
```

```python
final_rating.sample(5)
```

```python
final_rating.drop_duplicates(['user_id', 'title'], inplace=True)
```

```python
final_rating.shape
```

```python
final_rating.head()
```

```python
#final_rating.to_csv('final_rating.csv')
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```
